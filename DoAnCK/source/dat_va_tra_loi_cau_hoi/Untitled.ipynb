```python
def show_length_weight(order, max_length = 999999, max_weight = 9999999):
    order = data_df[data_df['Order'] == order]
    order = order[order['Length'] <= max_length]
    order = order[order['Weight'] <= max_weight]
    plt.scatter(x = order['Length'], y = order['Weight'])
```

```python
show_length_weight('Passeriformes', max_weight = 5)
```
