## Tính xem tình trạng của mỗi cột dữ liệu hiện tại như thế nào?

### Với mỗi cột ta sẽ tính:
- Tỉ lệ mất mát: missing_ratio
- Số lượng giá trị khác nhau: num_values
- Số lần xuất hiện của từng giá trị: value

Viết hàm "col_info" để tính và lưu kết quả vào mảng "cat_col_info_df"

```python
def col_info(col, array):
    temp = pd.DataFrame(array[col])
    temp = temp.dropna()
    temp = np.array(temp[col])
    data = []
    # missing_ratio
    data.append(round((100 - (len(temp) / len(array[col]) * 100)), 1))
    # num_values
    temp = np.unique(temp, return_counts=True)
    data.append(len(temp[0]))
    # value
    temp = pd.Series(data = temp[1], index = temp[0])
    data.append(temp)
    # num_values
    return data
```

```python
cat_col_info_df = pd.DataFrame(index = ['missing_ratio', 'num_values', 'value'])
for i in animal_df.columns:
    if animal_df[i].dtype not in [float, int]:
        cat_col_info_df[i] = col_info(i, animal_df)
        
pd.reset_option('display.max_colwidth')
cat_col_info_df
```

## Xử lý tách các cột có kiểu dictionary thành nhiều cột

Đầu tiên các cột dictionary đều có kiểu là string nên ta phải xử lý chuyển các cột trên thành dict

```python
for i in animal_df:
    if animal_df[i][0].find('{') != -1:
        animal_df[i] = animal_df[i].apply(eval)
```

Viết hàm "split_col" để split các cột có kiểu dữ liệu là dict thành nhiều cột

```python
def split_col(df):
    element = []
    for i in df:
        if type(df[i][0]) == dict: 
            element.append(i)
    df_split_animals_info = pd.DataFrame()
    for i in df:
        if i not in element:
            df_split_animals_info[i] = df[i]
        else:
            temp3 = df[i][0].items()
            temp3 = np.array(list(temp3), dtype = object)
            df_temp2 = pd.DataFrame(columns = temp3[:, 0])
            for rows in df[i]:
                if type(rows) != float:
                    temp3 = rows.items()
                else:
                    temp3 = []
                temp3 = np.array(list(temp3), dtype = object)
                if len(temp3) > 1: 
                    df_temp = pd.DataFrame(index = temp3[:, 0], data = temp3[:, 1]).T
                else:
                    df_temp = pd.DataFrame(columns = df_temp2.columns, index = [0])
                df_temp2 = pd.concat([df_temp2, df_temp], ignore_index=True)
            df_split_animals_info = df_split_animals_info.join(df_temp2)
    return df_split_animals_info
    
```

```python
df_split_animals_info = split_col(animal_df)
pd.set_option('display.max_columns', None)
df_split_animals_info
```

Sau khi split dữ liệu có số dòng là 3237 và số cột là 37, tăng lên 15 cột

Nhưng nhìn lại dữ liệu vẫn còn cột là dictionary như "Geograpy", như vậy ta phải split thêm 1 lần nữa

### Chuyển cột Top speed sang kiểu số

```python
def to_float(element):
    try:
        return float(element)
    except ValueError:
        return np.nan
```

```python
def trans_speed(speed):
    if pd.isna(speed): return np.nan
    speed = [speed[:-4].strip(), speed[-4:].strip()]
    if speed[1] == 'km/h':
        x = 1
    elif speed[1] == 'mph':
        x = 1.61
    sp = speed[0].split('-')
    if len(sp) == 1:
        return float(sp[0]) * x
    else:
        return (float(sp[0]) + float(sp[1])) / 2 * x
```

```python
data_df['Top speed'] = data_df['Top speed'].apply(trans_speed)
```

### Chuyển cột life_span sang float

```python
def trans_life_span(life):
    if pd.isna(life): return np.nan
    life = str(life)
    life = life.upper()
    life = life.replace('UPPER', '')
    life = life.replace('ABOVE', '')
    life = life.replace('OVER','')
    life = life.replace('mos', ' MONTHS ')
    life = life.replace('yr', ' YEARS ')
    life = life.replace('YRS', ' YEARS ')
    life = life.replace('YEARS', ' YEARS ')
    life = life.split('-')
    low = to_float(life[0])
    high = to_float(life[-1].split()[0].strip())
    x = life[-1].split()[-1].strip()
    if np.isnan(low):
        low = high
    t = 1.0
    if x == 'MONTHS':
        t = 1 / 12
    elif x == 'YEARS':
        t = 1.0
    low = low * t
    high = high * t
    return (low, high)
```

```python
data_df['Life span'] = data_df['Life span'].apply(trans_life_span)
#data_df['Life span'].dropna()
```

## Thống kê các chỉ số cho các cột có kiểu dữ liệu là list

```python
def stat_list_col(col, array):
    temp = pd.DataFrame(array[col])
    temp = temp.dropna()
    temp = np.array(temp[col])
    data = []
    # missing_ratio
    data.append(round((100 - (len(temp) / len(array[col]) * 100)), 1))
    # num_values
    temp = np.hstack(temp)
    temp = list(map(str.strip, temp)) # Xử lý các dấu cách đầu và cuối mỗi chuỗi
    temp = np.unique(temp, return_counts=True)
    data.append(len(temp[0]))
    # value
    temp = pd.Series(data = temp[1], index = temp[0])
    data.append(temp)
    # num_values
    return data

#     total = data_df[col].tolist()
#     total = np.hstack(total)
#     value, times = np.unique(total, return_counts = True)
    
```

```python
list_col_info_df = pd.DataFrame(index = ['missing_ratio', 'num_values', 'value'])
for i in data_df:
    if type(data_df[i][0]) == list:
        list_col_info_df[i] = stat_list_col(i, data_df)
list_col_info_df
```
